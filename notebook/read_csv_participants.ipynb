```python
f = open(participantlist, 'r')
for lines in f.readlines():
    country = lines.split(',')[-1].rstrip('\n')
    
    try:
        cc = pycountry.countries.get(name=country)
        print(country)
    except KeyError:
        print("{0}: Wrong country name".format(country))
        
f.close()
```
