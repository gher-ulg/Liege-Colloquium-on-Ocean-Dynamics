```python
import os
import glob
import csv
import pandas
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 14})
```

```python
datafile = "../data/numParticipant.csv"
figdir = "/home/ctroupin/dox/Projects/LiegeCLQ50/"
```

```python
df = pandas.read_csv(datafile)
```

```python
df[0]
```

```python
yearlist = []
participants = []
countries = []
with open(datafile, "r") as df1:
    # read first line
    line = df1.readline()
    for lines in df1:
        linesplit = lines.rstrip().split(",")
        yearlist.append(int(linesplit[0]))
        
        if len(linesplit[1]) > 0:
            participants.append(int(lines.split(",")[1]))
        else:
            participants.append(0)
            
        if len(linesplit[4]) > 0:
            countries.append(int(lines.split(",")[4]))
        else:
            countries.append(0)
```

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(111)
ax1.bar(yearlist, participants, color=".2", width=.35)
ax1.set_ylabel("Number of\nparticipants", rotation=0, ha="right")
ax1.spines['top'].set_visible(False)

ax2 = ax1.twinx()
ax2.set_ylabel("Number of\ncountries", rotation=0, ha="left", color="#0A87E2")
ax2.bar(0.35 + np.array(yearlist), countries, color="#0A87E2", width=.35)
ax2.tick_params(axis='y', labelcolor="#0A87E2")
ax2.spines['top'].set_visible(False)

plt.xticks(np.arange(1965, 2020, 5))
plt.xlim(1969, 2018)
plt.savefig(os.path.join(figdir, "participants_time.png"), dpi=300, bbox_inches="tight")
plt.savefig(os.path.join(figdir, "participants_time.pdf"), dpi=300, bbox_inches="tight")

plt.show()
plt.close()
```
