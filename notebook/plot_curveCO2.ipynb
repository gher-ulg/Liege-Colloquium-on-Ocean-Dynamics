```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams.update({'font.size': 25})
```

```python
datadir = "/home/ctroupin/Data/"
figdir = "../latex/"
datafile = os.path.join(datadir, "co2_mm_mlo.txt")
```

```python
dataCO2 = np.loadtxt(datafile)
dataCO2.shape
```

```python
fig = plt.figure(figsize=(15, 6))
ax = plt.subplot(111)
plt.plot(dataCO2[:, 2], dataCO2[:, 4])
plt.yticks([300, 400], color=".5")
plt.xticks([1969, 2018], fontsize=40, color=".5")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('.75')
ax.spines['left'].set_color('.75')
plt.savefig(os.path.join(figdir, "timeseries.png"), 
            dpi=300, bbox_inches='tight',
            transparent=True)
plt.show()
plt.close()
```
